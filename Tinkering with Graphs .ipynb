```python
# Data
C19PlusMajorCardiovascularDiseases = [100, 200, 300, 400, 500, 600, 700, 800, 900,]
AgeGroup = ['<18', '18-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-84', '85+']
```

```python
# Creating the line graph
plt.plot(C19PlusMajorCardiovascularDiseases, AgeGroup, marker='o')
plt.title('Age Distribution by C19PlusMajorCardiovascularDiseases')
plt.xlabel('C19PlusMajorCardiovascularDiseases')
plt.ylabel('AgeGroup')
plt.grid(True)
plt.show()
```

```python
#lets make a bar graph 
```

```python
# Filter the data for only those who died from COVID-19 and had Major Cardiovascular Diseases
covid_major_cardiovascular = df[df['C19PlusMajorCardiovascularDiseases'] == 1]
```

```python
# Counting occurrences of each age group among those who died from COVID-19 and had Major Cardiovascular Diseases
age_group_counts = covid_major_cardiovascular['AgeGroup'].value_counts()
```

```python
# Extracting categories and values
categories = age_group_counts.index.tolist()
values = age_group_counts.tolist()
```

```python
# Creating the bar chart
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.bar(categories, values, color='salmon')
plt.title('Distribution of Age Groups among COVID-19 Deaths with Major Cardiovascular Diseases')
plt.xlabel('Age Groups')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
#lets make a histogram 
```

```python
# Filter the data for only those who died from COVID-19 and had Major Cardiovascular Diseases
covid_major_cardiovascular = df[df['C19PlusMajorCardiovascularDiseases'] == 1]
```

```python
# Convert AgeGroup to strings using .loc[]
covid_major_cardiovascular.loc[:, 'AgeGroup'] = covid_major_cardiovascular['AgeGroup'].astype(str)
```

```python
# Extracting ages
ages = covid_major_cardiovascular['AgeGroup']
```

```python
# Creating the histogram
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.hist(ages, bins=10, color='orange')  # Adjust the number of bins for better visualization
plt.title('Distribution of Age Groups among COVID-19 Deaths with Major Cardiovascular Diseases')
plt.xlabel('Age Groups')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
#now lets make a pie chart 
```

```python
# Filter the data for only those who died from COVID-19 and had Major Cardiovascular Diseases
covid_major_cardiovascular = df[df['C19PlusMajorCardiovascularDiseases'] == 1]
```

```python
# Counting occurrences of each age group
age_group_counts = covid_major_cardiovascular['AgeGroup'].value_counts()
```

```python
# Extracting labels and values for the pie chart
labels = age_group_counts.index.tolist()
values = age_group_counts.tolist()
```

```python
# Creating the pie chart
plt.figure(figsize=(8, 8))  # Adjust figure size if needed
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Age Groups among COVID-19 Deaths with Major Cardiovascular Diseases')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

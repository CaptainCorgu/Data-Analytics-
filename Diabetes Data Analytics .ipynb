```python
from scipy.stats import f_oneway
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
# Melt the DataFrame to long format, which statsmodels formulas require
df_long = pd.melt(df, var_name='conditions', value_name='values')

# Verify that 'values' column contains only numeric data because you have dirty data 
if not pd.api.types.is_numeric_dtype(df_long['values']):
    df_long['values'] = pd.to_numeric(df_long['values'], errors='coerce')
    
# Check for and handle any NaN values that could cause issues in fitting the model
df_long = df_long.dropna(subset=['values'])    

# Fit the model
model = ols('values ~ C(conditions)', data=df_long).fit()

# Perform the ANOVA
anova_results = sm.stats.anova_lm(model, typ=2)

# Print the ANOVA table
print("ANOVA Table:")
print(anova_results)

# Check for significant differences
if anova_results['PR(>F)'][0] < 0.05:
    print("\nThere are significant differences between the groups.")
    # Conduct post-hoc analysis with Tukey's HSD test
    post_hoc = pairwise_tukeyhsd(endog=df_long['values'], groups=df_long['conditions'], alpha=0.05)
    print("\nPost-Hoc (Tukey's HSD) Results:")
    print(post_hoc)
else:
    print("\nNo significant differences were found between the groups.")
```
